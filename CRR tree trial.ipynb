```python
import numpy as np
import pandas as pd
import sklearn as sk
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
CRR = pd.read_csv('OBU_fakedata.csv')
CRR
```

```python
CRR.columns[0:-1]
```

```python
x = CRR[CRR.columns[0:-1]]
x
```

```python
x = x.fillna(99)
x
```

```python
y = CRR[CRR.columns[-1]]
y
```

```python
data_dum = pd.get_dummies(x)
data_dum = pd.DataFrame(data_dum)
data_dum
```

```python
data_dum.to_csv('Dummy.csv')
```

```python
from sklearn.model_selection import train_test_split
x = data_dum
y = CRR[CRR.columns[-1]]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
# argument: max_depth=k
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=5)
```

```python
model.fit(x_train, y_train)
```

```python
plt.figure(figsize=(20,20))
tree.plot_tree(model) 
```

```python
fn = x.columns
cn = y.unique()
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (25,25), dpi=300)
tree.plot_tree(model,
               feature_names = fn, 
               class_names=cn,
               filled = True);
fig.savefig('tree.png')
```

```python
from sklearn.metrics import f1_score, recall_score, precision_score ,mean_squared_error, r2_score, accuracy_score, confusion_matrix
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
con_matrix = confusion_matrix(y_test, y_pred)
recall_score = recall_score(y_test, y_pred, average='weighted')
precision_score = precision_score(y_test, y_pred, average='weighted')
f1_score = f1_score(y_test, y_pred, average='weighted')
```
